```
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```
penguins = sb.load_dataset("penguins")
```

```
penguins.head()
```

```
fig, ax = plt.subplots(1, 3,figsize=(20,10))

#displots distributions
plt.grid()

sb.kdeplot(data=penguins, x="flipper_length_mm", hue="species", alpha=.8, multiple="stack",ax=ax[0])
ax[0].set(title='KDE PLOT')
ax[0].grid()
sb.histplot(data=penguins, x="flipper_length_mm", hue="species", shrink=.8, alpha=.8,ax=ax[1])
ax[1].set(title='Histogram PLOT')
sb.scatterplot(data=penguins,x="flipper_length_mm", y="bill_length_mm", hue="species",ax=ax[2])
ax[2].set(title='Scatterplot')


fig.tight_layout()

plt.show()
```

# Relationship or joint distribution of two variables

```
sb.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```
